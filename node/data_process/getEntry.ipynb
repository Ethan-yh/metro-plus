```python

import json

with open('../data/entry_data_ori.json', 'r', encoding='utf-8')as f:
    entry_data_ori = json.load(f)
    # print(entry_data_ori)

with open('../data/metro_data.json', 'r', encoding='utf-8') as f:
    metro_data = json.load(f)
    stations = metro_data['stations']

entry_data = {}

# entry_data['g23'] = 34
# print(entry_data)
notin_count = 0

def getStation(name):
    global notin_count
    for station in stations:
        if station['name'] == name:
            return station
    print(f'没有该站:{name}')
    notin_count = notin_count+1
    return None

for items in entry_data_ori:
    entry_data[str(items[0]['minutegroup'])] = []
    for item in items:
        station = getStation(item['name'])
        if not station:
            continue
        # print(station)
        record = {'station_id':station['id'], 'station_name':station['name'], 'entry':item['entry'], 'minutegroup': item['minutegroup']}
        entry_data[str(items[0]['minutegroup'])].append(record)

print(f'{notin_count}站未找到')
with open('../data/entry_data.json', 'w') as f:
    json.dump(entry_data, f, ensure_ascii=False)
```
